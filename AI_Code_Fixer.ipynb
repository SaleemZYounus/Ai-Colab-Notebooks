```python
from google import genai
import os

def self_healing(code):
    # Set your API key free w
    os.environ["Unlocker"] = "API_KEY"

    # Initialize Gemini Client
    client = genai.Client(api_key=os.environ["Unlocker"])

    # Gemini API prompt
    question = "fix the following code, only output the fixed code, the output should be formated in propper python format, add a comment at the end explaining the fix briefly, make sure its properly commented out"
    question = question + ("Code: ") + code
    # Calling Gemini API 'gemini-2.5-flash' model to generate content
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=question,
    )
    output = response.text
    return (output)
```

```python
with open('file.py', 'r') as f:
    code = f.read()

newfile = self_healing(code)[10:-4]
```

```python
with open('updated.py', 'w') as file:
    file.write(newfile)
```
